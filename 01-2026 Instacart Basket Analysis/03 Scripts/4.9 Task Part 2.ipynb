Step 3: Create a histogram of the “order_hour_of_day” column

```python
# Creating a histogram 
hist = df_final_dataframe['order_hour_of_day'].plot.hist(bins = 24)
plt.tight_layout()
```

Most orders are placed between 8am to 4pm. After that time, order frequency starts to decline. Most customers place their orders between late morning and early afternoon with very little orders placed during the night.

Stept 4: Create a bar chart from the “loyalty_flag” column

```python
# reating a bar chart
bar = df_final_dataframe['loyalty_flag'].value_counts().plot.bar()
plt.tight_layout()
```

```python
df_final_dataframe['loyalty_flag'].value_counts().sort_index()
```

Regular customers make up the largest group, with 473,070 users. This suggests there is strong potential to convert them into loyal customers through rewards or personalized offers. Also, There are many new customers joining, but they need proper onboarding to encourage them to order again.
