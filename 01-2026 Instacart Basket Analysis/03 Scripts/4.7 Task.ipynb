Creating price_label column

```python
# To avoid memory issues i'll work with a subset
df = df_ords_prods_merge[:1000000]
```

```python
df_ords_prods_merge.shape
```

```python
# Defining a function for "price label"
def price_label(row): 
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else:
        return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts (dropna = False)
```

```python
# Checking what the most expensive product within the subset is
df['prices'].max()
```

```python
# If-Statements with the loc() Function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna=False)
```

```python
# Applying If- Statements to full dataframe
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

Creating busiest_day column

```python
# Frequency of orders_day_of_the_week
df_ords_prods_merge['orders_day_of_week'].value_counts()
```

```python
#  Using a for-loop to determine whether orders are on "busiest day" (0 = Saturday), "least busy" (4 = Wednesday), or "regularly busy" (other days of the week)
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Printing the results
result
```

```python
# Creating a busiest_day column within the dataframe and setting it equal to result
df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

Step 2

```python
# Update busiest_day column to busiest_days
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        result.append("Busiest days")
    elif value == 4 or value == 3:
        result.append("Slowest days")
    else:
        result.append("Regularly busy")
```

```python
# Printing the results
result
```

Step 3

```python
# Checking the values of new busiest_days column
df_ords_prods_merge['busiest_days'] = result
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna=False)
```

Most orders occur on regularly busy days, followed by busiest days, while slowest days have the fewest orders. This shows that orders usually happen on moderately to highly active days.

Step 4

```python
# Create a new column, busiest_period_of_day
# Check value counts in "order_hour_of_day" column.
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

Most orders = 10, 11, 14, 15, 13, 12, 16, and 9; Fewest orders = 23, 6, 0, 1, 5, 2, 4, and 3; Average orders = all remaining values in order_hour_of_day.

```python
# Creating a for-loop if statement labeling periods of time as Most orders, Average orders, and Fewest orders.
result = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]: 
      result.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]: 
      result.append("Fewest orders")
  else: result.append("Average orders")
```

```python
# Printing the results
result
```

```python
# Creating new column "busiest_period_of_day" in ords_prods_merge.
df_ords_prods_merge['busiest_period_of_day'] = result
```

Step 5

```python
# Printing the frequency of busiest_period_of_day column
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
